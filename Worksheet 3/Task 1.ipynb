# Task 1: Implementation of MCP Neurons

In this task, we implement the McCulloch-Pitts (MCP) neuron model to simulate basic Boolean functions like **AND** and **OR**.

```python
def MCP_Neurons_AND(X1, X2, T):
    """
    This functions implements basic AND operations with MCP Neuron for two inputs.
    Arguments:
    Inputs:
    X1 (1 nd array): An array of binary values.
    X2 (1 nd array): An array of binary values.
    Output:
    state_neuron(1D-list): An state of neuron 1 0r 0 for the particular inputs.
    """
    assert len(X1) == len(X2)
    
    state_neuron = []
    # Perform an element wise addition of two input arrays and compare with Threshold
    for x1, x2 in zip(X1, X2):
        if (x1 + x2) >= T:
            state_neuron.append(1)
        else:
            state_neuron.append(0)
            
    return state_neuron
```

```python
# Example usage for MCP_Neurons_AND function
X1 = [0, 0, 1, 1]
X2 = [0, 1, 0, 1]
T = 2 # Threshold value for AND (requires both inputs to be 1)

# Call the MCP_Neurons_AND function
result = MCP_Neurons_AND(X1, X2, T)

# Print the result
print(f"Output of AND gate for inputs {X1} and {X2} with threshold {T}: {result}")
```

### 3.2.1 Answer the Following Question:

**Question - 1: List out all the limitations of MCP - Neurons.**

1. **Fixed Weights:** The weights and thresholds are handcrafted and fixed. There is no automated learning algorithm to adjust them.
2. **Binary Inputs/Outputs:** It only handles binary (0/1) data, which limits its use for continuous values.
3. **Linear Separability:** A single MCP neuron can only solve linearly separable problems. It cannot solve non-linear problems like XOR.
4. **No Bias Term:** Unlike the Perceptron, it doesn't have an explicit learnable bias; the threshold must be manually tuned.

**Question - 2: Think if you can develop a logic to solve for XOR function using MCP Neuron.**

To solve the XOR function, we would need a multi-layer network of MCP neurons. XOR can be broken down into simpler logical components:
`XOR(x1, x2) = (x1 OR x2) AND NOT (x1 AND x2)`

In terms of MCP neurons, this requires:
1. One neuron for the **OR** operation.
2. One neuron for the **AND** operation.
3. An inhibitory connection (representing **NOT**) from the AND neuron's output into a final decision neuron that also receives input from the OR neuron.

A single MCP neuron cannot solve XOR because the data is not linearly separable.
