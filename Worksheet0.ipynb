### **Task - 1:**
Create a Python program that converts between different units of measurement.
*   The program should:
    1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
    2. Ask the user to input the value to be converted.
    3. Perform the conversion and display the result.
    4. Handle potential errors, such as invalid input or unsupported conversion types.
*   Requirements:
    1. **Functions:** Define at least one function to perform the conversion.
    2. **Error Handling:** Use try-except blocks to handle invalid input (e.g., non-numeric values).
    3. **User Input:** Prompt the user to select the conversion type and input the value.
    4. **Docstrings:** Include a docstring in your function to describe its purpose, parameters, and return value.
*   Conversion Options:
    1. **Length:**
        - Convert meters (m) to feet (ft).
        - Convert feet (ft) to meters (m).
    2. **Weight:**
        - Convert kilograms (kg) to pounds (lbs).
        - Convert pounds (lbs) to kilograms (kg).
    3. **Volume:**
        - Convert liters (L) to gallons (gal).
        - Convert gallons (gal) to liters (L).

```python
def convert_units(conversion_type, value):
  """
  Converts units between length, weight, and volume.

  Parameters:
    conversion_type (int): The selected conversion option.
    value (float): The numeric value to convert.

  Returns:
    float: The converted value.
  """

  # Length conversions
  if conversion_type == 1:          # meters to feet
    return value * 3.28084
  elif conversion_type == 2:        # feet to meters
    return value / 3.28084

  # Weight conversions
  elif conversion_type == 3:        # kilograms to pounds
    return value * 2.20462
  elif conversion_type == 4:        # pounds to kilograms
    return value / 2.20462

  # Volume conversions
  elif conversion_type == 5:        # liters to gallons
    return value * 0.264172
  elif conversion_type == 6:        # gallons to liters
    return value / 0.264172

  else:
    raise ValueError("Unsupported conversion type selected.")

try:
  print("Unit Conversion Program")
  print("Choose a conversion type:")
  print("1. Meters (m) to Feet (ft)")
  print("2. Feet (ft) to Meters (m)")
  print("3. Kilograms (kg) to Pounds (lbs)")
  print("4. Pounds (lbs) to Kilograms (kg)")
  print("5. Liters (l) to Gallons (gal)")
  print("6. Gallons (gal) to Liters (l)")

  choice = int(input("Enter your choice (1-6): "))
  value = float(input("Enter the value to convert: "))

  result = convert_units(choice, value)

  print("Converted value:", result)

except ValueError as e:
  print("Error:", e)

except Exception:
  print("Invalid input. Please enter numeric values only.")
```

### **Task - 2:**
Create a Python program that performs various mathematical operations on a list of numbers.
*   The Program should:
    1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum of the numbers).
    2. Ask the user to input a list of numbers (separated by spaces).
    3. Perform the selected operation and display the result.
    4. Handle potential errors, such as invalid input or empty lists.
*   **Requirements:**
    1. **Functions:** Define at least one function for each operation (sum, average, maximum, minimum).
    2. **Error Handling:** Use try-except blocks to handle invalid input (e.g., non-numeric values or empty lists).
    3. **User Input:** Prompt the user to select the operation and input the list of numbers.
    4. **Docstrings:** Include a docstring in each function to describe its purpose, parameters, and return value.

```python
def calculate_sum(numbers):
  """
  Returns the sum of a list of numbers.

  Parameters:
    numbers (list of float): List of numeric values.

  Returns:
    float: Sum of the numbers.
  """
  total = 0
  for number in numbers:
    total += number
  return total

def calculate_average(numbers):
  """
  Returns the average of a list of numbers.

  Parameters:
    numbers (list of float): List of numeric values.

  Returns:
    float: Average of the numbers.
  """
  if len(numbers) == 0:
    raise ValueError("Cannot calculate average of an empty list.")
  return calculate_sum(numbers) / len(numbers)

def find_maximum(numbers):
  """
  Returns the maximum value in a list of numbers.

  Parameters:
    numbers (list of float): List of numeric values.

  Returns:
    float: Maximum value.
  """
  if len(numbers) == 0:
    raise ValueError("Cannot find maximum of an empty list.")

  maximum = numbers[0]
  for number in numbers:
    if number > maximum:
      maximum = number
  return maximum

def find_minimum(numbers):
  """
  Returns the minimum value in a list of numbers.

  Parameters:
    numbers (list of float): List of numeric values.

  Returns:
    float: Minimum value.
  """
  if len(numbers) == 0:
    raise ValueError("Cannot find minimum of an empty list.")

  minimum = numbers[0]
  for number in numbers:
    if number < minimum:
      minimum = number
  return minimum

try:
  print("Mathematical Operations Program")
  print("Choose an operation:")
  print("1. Sum")
  print("2. Average")
  print("3. Maximum")
  print("4. Minimum")

  choice = int(input("Enter your choice (1-4): "))

  user_input = input("Enter numbers separated by spaces: ")

  numbers = [float(x) for x in user_input.split()]

  if len(numbers) == 0:
    raise ValueError("The list cannot be empty.")

  if choice == 1:
    result = calculate_sum(numbers)
    print("Sum:", result)

  elif choice == 2:
    result = calculate_average(numbers)
    print("Average:", result)

  elif choice == 3:
    result = find_maximum(numbers)
    print("Maximum:", result)

  elif choice == 4:
    result = find_minimum(numbers)
    print("Minimum:", result)

  else:
    raise ValueError("Invalid operation selected.")

except ValueError as e:
  print("Error:", e)

except Exception:
  print("Invalid input. Please enter numbers correctly.")
```

### **1. Flatten a Nested List:**
Write a Python function that takes a nested list and flattens it into a single list, where all the elements are in a single dimension.
*   **Requirements:**
    *   Define a function flatten(lst) that takes a nested list lst and returns a flattened version of the list.
    *   Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(lst):
  """
  Flattens a nested list into a single list.

  Parameters:
    lst (list): A nested list.

  Returns:
    list: A flattened list.
  """
  flat_list = []
  for sublist in lst:
    for item in sublist:
      flat_list.append(item)
  return flat_list

numbers = [[1, 2], [3, 4], [5]]
result = flatten(numbers)
print(result)
```

    *   Example: For the input lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] with indices = [1, 2], the output should be 6.

```python
def access_nested_element(lst, indices):
  """
  Returns a specific element from a nested list based on a list of indices.

  Parameters:
    lst (list): A nested list.
    indices (list): A list of indices showing the path.

  Returns:
    element at the specified nested position.
  """
  element = lst
  for index in indices:
    element = element[index]
  return element

lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
result = access_nested_element(lst, [1, 2])
print(result)
```

### **3. Sum of All Elements in a Nested List:**
Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).
*   **Requirements:**
    *   Define a function sum nested(lst) that takes a nested list lst and returns the sum of all the elements.
    *   Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 21.

```python
def sum_nested(lst):
  """
  Returns the sum of all numbers in a nested list (any depth).

  Parameters:
    lst (list): A nested list containing numbers.

  Returns:
    int/float: Sum of all elements.
  """
  total = 0
  for item in lst:
    if isinstance(item, list):
      total += sum_nested(item)
    else:
      total += item
  return total

numbers = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(numbers)
print(result)
```

### **4. Remove Specific Element from a Nested List:**
Write a Python function that removes all occurrences of a specific element from a nested list.
*   **Requirements:**
    *   Define a function remove element(lst, elem) that removes elem from lst and returns the modified list.
    *   Example: For the input lst = [[1, 2], [3, 2], [4, 5]] and elem = 2, the output should be [[1], [3], [4, 5]].

```python
def remove_element(lst, elem):
  """
  Removes all occurrences of elem from a nested list.

  Parameters:
    lst (list): A nested list.
    elem: Element to remove.

  Returns:
    list: Modified list with elem removed.
  """
  new_list = []

  for item in lst:
    if isinstance(item, list):
      new_list.append(remove_element(item, elem))
    else:
      if item != elem:
        new_list.append(item)

  return new_list

lst = [[1, 2], [3, 2], [4, 5]]
result = remove_element(lst, 2)
print(result)
```

### **5. Find the Maximum Element in a Nested List:**
Write a Python function that finds the maximum element in a nested list (regardless of depth).
*   **Requirements:**
    *   Define a function find max(lst) that takes a nested list lst and returns the maximum element.
    *   Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_max(lst):
  """
  Finds the maximum element in a nested list (any depth).

  Parameters:
    lst (list): A nested list containing numbers.

  Returns:
    int/float: Maximum element in the list.
  """
  maximum = None

  for item in lst:
    if isinstance(item, list):
      value = find_max(item)
    else:
      value = item

    if maximum is None or value > maximum:
        maximum = value

  return maximum

numbers = [[1, 2], [3, [4, 5]], 6]
result = find_max(numbers)
print(result)
```

### **6. Count Occurrences of an Element in a Nested List:**
Write a Python function that counts how many times a specific element appears in a nested list.
*   **Requirements:**
    *   Define a function count occurrences(lst, elem) that counts the occurrences of elem in the nested list lst.
    *   Example: For the input lst = [[1, 2], [2, 3], [2, 4]] and elem = 2, the output should be 3.

```python
def count_occurrences(lst, elem):
  """
  Counts how many times elem appears in a nested list.

  Parameters:
    lst (list): A nested list.
    elem: Element to count.

  Returns:
    int: Number of occurrences of elem.
  """
  count = 0

  for item in lst:
    if isinstance(item, list):
      count += count_occurrences(item, elem)
    else:
      if item == elem:
        count += 1

  return count

lst = [[1, 2], [2, 3], [2, 4]]
result = count_occurrences(lst, 2)
print(result)
```

### **7. Flatten a List of Lists of Lists:**
Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.
*   **Requirements:**
    *   Define a function deep flatten(lst) that takes a deeply nested list lst and returns a single flattened list.
    *   Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be [1, 2, 3, 4, 5, 6, 7, 8].

```python
def deep_flatten(lst):
  """
  Flattens a deeply nested list into a single list.

  Parameters:
    lst (list): A deeply nested list.

  Returns:
    list: A completely flattened list.
  """
  flat_list = []

  for item in lst:
    if isinstance(item, list):
      flat_list.extend(deep_flatten(item))
    else:
      flat_list.append(item)

  return flat_list

lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(lst)
print(result)
```

### **8. Nested List Average:**
Write a Python function that calculates the average of all elements in a nested list.
*   **Requirements:**
    *   Define a function average nested(lst) that takes a nested list lst and returns the average of all the elements.
    *   Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def average_nested(lst):
  """
  Calculates the average of all numbers in a nested list.

  Parameters:
    lst (list): A nested list containing numbers.

  Returns:
    float: Average of all elements.
  """

  def helper(sublist):
    total = 0
    count = 0

    for item in sublist:
      if isinstance(item, list):
        sub_total, sub_count = helper(item)
        total += sub_total
        count += sub_count
      else:
        total += item
        count += 1

    return total, count

  total_sum, total_count = helper(lst)
  return total_sum / total_count

lst = [[1, 2], [3, 4], [5, 6]]
result = average_nested(lst)
print(result)
```

*   Using **Python Lists**, perform element-wise addition of two lists of size 1, 000, 000. Measure and Print the time taken for this operation.
*   Using **Numpy Arrays**, Repeat the calculation and measure and print the time taken for this operation.

```python
import numpy as np
import time

# 1. Using Python Lists
size = 1_000_000
list1 = list(range(size))
list2 = list(range(size))

start_time = time.time()
list_sum = [list1[i] + list2[i] for i in range(size)]
end_time = time.time()
python_time = end_time - start_time
print("Time taken for element-wise addition using Python lists:", python_time, "seconds\n")

# 2. Using NumPy Arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

start_time = time.time()
array_sum = arr1 + arr2
end_time = time.time()
numpy_time = end_time - start_time
print("Time taken for element-wise addition using NumPy arrays:", numpy_time, "seconds\n")

# Comparison
speedup = python_time / numpy_time
print(f"NumPy is approximately {speedup:.2f} times faster than Python lists for element-wise addition.")
```

### **4. Matrix Multiplication**
*   Using **Python lists**, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.
*   Using **NumPy arrays**, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.


```python
import numpy as np
import time

# Matrix size
size = 1000

# 1. Using Python Lists
# Create two 1000x1000 matrices filled with 1s
matrix1 = [[1 for _ in range(size)] for _ in range(size)]
matrix2 = [[1 for _ in range(size)] for _ in range(size)]

start_time = time.time()

# Initialize result matrix
result = [[0 for _ in range(size)] for _ in range(size)]

# Perform matrix multiplication (triple nested loop)
for i in range(size):
    for j in range(size):
        for k in range(size):
            result[i][j] += matrix1[i][k] * matrix2[k][j]

end_time = time.time()
python_time = end_time - start_time
print("Time taken for matrix multiplication using Python lists:", python_time, "seconds\n")

# 2. Using NumPy Arrays
arr1 = np.ones((size, size))
arr2 = np.ones((size, size))

start_time = time.time()
result_np = np.dot(arr1, arr2)  # or arr1 @ arr2
end_time = time.time()
numpy_time = end_time - start_time
print("Time taken for matrix multiplication using NumPy arrays:", numpy_time, "seconds\n")

# Comparison
speedup = python_time / numpy_time
print(f"NumPy is approximately {speedup:.2f} times faster than Python lists for matrix multiplication.")
```
